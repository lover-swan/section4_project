```python
import pandas as pd
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/section4_data_v2.csv', index_col=0)
df1
```

```python
df1['token_len'] = df1['tokens'].apply(lambda x: len(eval(x)))
df1
```

```python
df1 = df1[df1['token_len']>2].reset_index(drop=True)
df1
```

```python
import re
```

```python
def clear(text):

  tokens = re.sub(r'[^a-zA-Z ]','',text)
  tokens = tokens.lower()

  return tokens
```

```python
df1['sentence']=df1['tokens'].apply(clear)
df1
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vect = TfidfVectorizer(stop_words='english', ngram_range=(1,3), min_df=4,max_df=0.7, max_features=20000)

dtm_tfidf_amazon = tfidf_vect.fit_transform(df1['sentence'])

dtm_tfidf_amazon = pd.DataFrame(dtm_tfidf_amazon.todense(), columns=tfidf_vect.get_feature_names())
dtm_tfidf_amazon
```

```python
dtm_tfidf_amazon.to_csv('/content/drive/MyDrive/section4_vector_data.csv')
```

```python
df = pd.read_csv('/content/drive/MyDrive/section4_vector_data.csv', index_col=0)
df
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
nn = NearestNeighbors(n_neighbors=10, algorithm='kd_tree')
nn.fit(dtm_tfidf_amazon)
```

```python
qusetion1 = int(input('검색할 리뷰 번호를 입력하세요 '))
index1 = nn.kneighbors([dtm_tfidf_amazon.iloc[qusetion1]])[1]
df1.loc[index1[0]][{'asin','reviewText','summary'}]
```

```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('all-mpnet-base-v2')


sentences = df1['sentence']

paraphrases = util.paraphrase_mining(model, sentences,corpus_chunk_size=len(sentences),show_progress_bar=True,top_k=9,batch_size=64)
```

```python
len(paraphrases)
```

```python
dict = {}

for score, i, j in paraphrases:
  if i not in dict.keys():
    dict[i] = [j]
  else:
    dict[i].append(j)
```

```python
len(dict)
```

```python
search_df = pd.DataFrame(data=zip(dict.keys(),dict.values()),columns=['key','value']).sort_values(by='key').reset_index(drop=True)
search_df
```

```python
qusetion = int(input('검색할 리뷰 번호를 입력하세요 '))
a = list(search_df[search_df['key']==qusetion]['value'])
df1.loc[a[0]][{'asin','reviewText','summary'}]
```
